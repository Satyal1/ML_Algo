# Importing Required Libraries 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('Salary_Data.CSV')  # Importing data set
```

```python
df.info()         # checking imported data
```

```python
x=df.iloc[:,:-1].values   # If there is no null values assinging x& y Values
y=df.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split      # dividing data sets to Train & test for model creation
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.5,random_state=20)
```

```python
from sklearn.linear_model import LinearRegression  # importing LR for prediction of x_test values
reg=LinearRegression()
```

```python
reg.fit(x_train,y_train)
y_pred=reg.predict(x_test)
```

```python
plt.scatter(x_train,y_train,c='red')              # plottiing test & train datas
plt.plot(x_train,reg.predict(x_train),c='blue')
plt.title('Salary vs year of experiance(training set)')
plt.xlabel('years of experiance')
plt.ylabel('salary')
plt.show()
```

```python
plt.scatter(x_test,y_test,c='red')
plt.plot(x_train,reg.predict(x_train),c='blue')
plt.title('Salary vs year of experiance(testing set)')
plt.xlabel('years of experiance')
plt.ylabel('salary')
plt.show()
```

```python
from sklearn.metrics import r2_score   # importing r2_score to check model score
```

```python
r2_score(y_test,y_pred)
```
